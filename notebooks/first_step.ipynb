# Conditional structures

## `IF` and `ELSE`

`if` and `else` are two conditional structures. The `if` will execute the command block if the condition is **true**. The `else` is a case where the `if` conditional is **false**.

The `if` is a keyword in Python that means “if”. It is used to form a conditional structure, which allows you to check whether a certain condition is true or false and then execute a specific block of code depending on the result of the check. The syntax for using `if` is:



```
if condition:
    # do something
```


````python
if 2>7:
  print('condition true')
  print()
print('out of block')
````

In Python, `else` is used in conjunction with the `if` keyword to form a conditional structure. The syntax for using `else` is:

```
if condition:
  # code if true
else:
  # code if false
```

The `else` is executed when the condition checked by the `if` evaluates to `False`.

---
**Situation:**

We will receive a student's GPA and we need an algorithm that performs the analysis and decides whether that student is **Passed** or Failed, displaying a message of the result. To pass, the average must be equal to or greater than 6.0.

````python
media = float(input('Enter average: '))

if media >= 6.0:
  print('Passed(a)')
else:
  print('Fail(a)')
````

Now our educational institution has released an official notice that people who have an average between 4.0 and 6.0 can take **Recovery** courses during the vacations in order to recover their grade.

So we can now make a set of `if`s so that we can structure this new condition.

````python
media = float(input('Enter average: '))

if media >= 6.0:
  print('Passed(a)')
if 6.0 > media >= 4.0:
  print('Recovery')
if media < 4.0:
  print('Fail(a)')
````

Note that in cases with 3 situations like this, we need to define our conditions well. Because the construction was made with `else` at the end, it will only consider some conditional to be **false** and may result in two (or more) executions.

For example:

````python
media = float(input('Enter average: '))

if media >= 6.0:
  print('Passed(a)')
if 6.0 > media >= 4.0:
  print('Recovery')
else:
  print('Fail(a)')
````

## `ELIF`

The `elif` is a Python keyword that means “else, if” and can be considered a union of the `else` and an `if`. It is used in conjunction with the `if` keyword to form a chained conditional structure.


The syntax for using `elif` is:

```
if condition1:
    # do something
elif condition2:
    # do something else
elif condition3:
    # do something else
else:
    # do something else
```

The `elif` allows you to check several conditions in a chained fashion, saving space in your code. If the first condition evaluates to `False`, the Python interpreter will evaluate the next condition in the `elif`. This will continue until a condition evaluates to `True` or until `else` is reached. If none of the conditions evaluate to `True`, execution of the `else` code will begin.

Let's use the same case as before:

````python
media = float(input('Enter average: '))

if media >= 6.0:
  print('Passed(a)')
elif 6.0 > media >= 4.0:
  print('Recovery')
else:
  print('Fail')
````

## Operators

When constructing commands, we sometimes need to elaborate on the conditional expression, requiring some logical operators to be integrated.

### `AND`, `OR`, `NOT`

The logical operators `and`, `or` and `not` are used to combine logical expressions in Python. They are often used in conjunction with `if` to create more complex conditional structures.

- `AND` is used to check whether two conditions are true. The logical expression¹ `x and y` is evaluated as `True` only if **both conditions `x` and `y` are true**, and as `False` otherwise.

- `OR` is used to check if at least one of the conditions is true. The logical expression `x or y` is evaluated as `True` **if at least one of the conditions `x` or `y` is true**, and as `False` if both are false.

- `NOT` is used to **deny a condition**. The logical expression not x is evaluated as True if condition x is false, and as False if condition x is true.

¹ A logical expression is a statement that can be evaluated as true or false. It is made up of logical operands² and logical operators³, which are used to combine several logical expressions into a single expression.

² Logical operands are the elements that are compared or evaluated in a logical expression. They are usually true or false values, but can also be more complex logical expressions. In Python, the logical operands are the values `True` and `False`.

Logical operators are symbols or keywords that are used to combine several logical expressions into a single expression. In Python, the logical operators are `and`, `or` and `not`, as well as the keywords `if`, `elif` and `else`.

````python
t1 = t2 = True
f1 = f2 = False
````

````python
if t1 and f2:
  print('true expression')
else:
  print('false expression')
````

````python
if t1 or f2:
  print('true expression')
else:
  print('false expression')
````

````python
if not f1:
  print('true expression')
else:
  print('false expression')
````

# Repetition structures

When we have to execute the same block of commands several times, it's not very interesting to do it by hand.

Let's imagine the situation where we have to collect and print the average of two grades from **3 students**:

````python
note_1 = float(input('Type to 1st note: '))
note_2 = float(input('Type a 2nd note: '))

print(f'Media: {(note_1+note_2)/2}')
note_1 = float(input('Type a 1st note: '))
note_2 = float(input('Type a 2nd note: '))

print(f'Media: {(note_1+note_2)/2}')
note_1 = float(input('Type a 1st note: '))
note_2 = float(input('Type a 2nd note: '))

print(f'Media: {(note_1+note_2)/2}')
````

Now let's imagine a situation where there are not just 3 students, but 100 students. It wouldn't be interesting to repeat the same code 100 times, but to **run the same code 100 times**.

We can build this repetition with repeat loops!

## `WHILE`

The `while` loop is a repetition control structure in Python that allows you to execute a block of code repeatedly while a certain condition is true. Its structure is:



```
while condition:
    # code block
```

Let's build an example with a counter from 1 to 10.

````python
counter = 1
while counter <= 10:
  print(counter)
  counter += 1
````

Now let's collect the grades and averages for each student in the while. We'll do an example with 3 averages.

````python
counter = 1
while counter <= 3:
  note_1 = float(input('Enter 1st note: '))
  grade_2 = float(input('Enter 2nd grade: '))

  print(f'Average: {(note_1+note_2)/2}')
  counter += 1
````

## `FOR`

The `for` loop is a type of flow control structure in Python that allows you to iterate over a set of elements. Its structure is:



```
for element in set:
    # code to be executed for each element
```

The for loop iterates over each element of the specified set and executes the code block inside the loop for each element. When the loop reaches the end of the set, it is interrupted and the program continues executing after the loop.


The set can be generated with the function [`range()`](https://docs.python.org/3/library/functions.html#func-range). This is a function capable of generating a sequence of integers. The structure of this function is:

```
range(start, end, step)
```

According to the documentation, `range()` generates a sequence of integers from the value of the `start` parameter to the value of the `end` parameter, according to the value of the `step` parameter. If `start` is not specified, the default value is 0. If `step` is not specified, the default value is 1.



Let's do the same `while` counter now with `for`.

````python
for counter in range(1,11):
  print(counter)
````

````python
for counter in range(1,4):
  grade_1 = float(input('Enter 1st grade: '))
  grade_2 = float(input('Enter 2nd grade: '))

  print(f'Average: {(note_1+note_2)/2}')
````

## Dictionary

Dictionaries are a type of data structure that store *key-value* pairs. They are delimited by braces `{}` and the *key-value* pairs are separated by commas.

```
dictionary = {key: value}
```

The **key** is a unique element that identifies a value in the dictionary, while the **value** is the item that is stored for the key. Keys and values can be of **any data type**.

Dictionaries are useful for storing and accessing data in an organized and fast way. They are a type of element set in Python, as they store a collection of items.

````python
dictionary = {'key_1':1,
              'key_2':2}
dictionary
````
